```python
# %% [markdown]
# # Разведочный анализ данных (EDA)
# ## Датасет: Bank Churn Competition

# %% [markdown]
# ### 1. Загрузка необходимых библиотек

# %%
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Настройка стилей для графиков
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
%matplotlib inline

# %% [markdown]
# ### 2. Загрузка данных

# %%
# Загрузка данных
df = pd.read_csv('D:\project-ai-Bank-Score\data\Churn_Modelling.csv')  # или используйте sample для теста
# df = pd.read_csv('../data/sample/sample_train.csv')  # для быстрого тестирования

print(f"Размер датасета: {df.shape}")
print(f"Количество признаков: {df.shape[1]}")
print(f"Количество записей: {df.shape[0]}")

# %%
# Первые 5 строк
df.head()

# %%
# Основная информация о данных
df.info()

# %% [markdown]
# ### 3. Проверка пропущенных значений

# %%
# Проверка пропусков
missing_values = df.isnull().sum()
missing_percent = (missing_values / len(df)) * 100

missing_df = pd.DataFrame({
    'Пропущено': missing_values,
    'Процент': missing_percent
})
print(missing_df[missing_df['Пропущено'] > 0])

# Визуализация пропусков
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, yticklabels=False, cmap='viridis')
plt.title('Карта пропущенных значений')
plt.show()

# %% [markdown]
# ### 4. Статистическое описание данных

# %%
# Описательная статистика для числовых признаков
df.describe()

# %%
# Описательная статистика для категориальных признаков
categorical_cols = df.select_dtypes(include=['object']).columns
df[categorical_cols].describe()

# %% [markdown]
# ### 5. Анализ целевой переменной (Exited)

# %%
# Распределение целевой переменной
target_dist = df['Exited'].value_counts()
target_percent = df['Exited'].value_counts(normalize=True) * 100

print("Распределение целевой переменной:")
print(f"0 (Остались): {target_dist[0]} ({target_percent[0]:.2f}%)")
print(f"1 (Ушли): {target_dist[1]} ({target_percent[1]:.2f}%)")

# Визуализация
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Count plot
sns.countplot(data=df, x='Exited', ax=axes[0], palette=['#2ecc71', '#e74c3c'])
axes[0].set_title('Распределение оттока клиентов')
axes[0].set_xlabel('Exited (0 - остался, 1 - ушел)')
axes[0].set_ylabel('Количество')

# Pie chart
colors = ['#2ecc71', '#e74c3c']
axes[1].pie(target_dist.values, labels=['Остались', 'Ушли'], 
            autopct='%1.1f%%', colors=colors, startangle=90)
axes[1].set_title('Доля оттока клиентов')

plt.tight_layout()
plt.show()

# %% [markdown]
# ### 6. Анализ числовых признаков

# %%
# Выделим числовые признаки (исключая идентификаторы)
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
numeric_cols.remove('RowNumber')
numeric_cols.remove('CustomerId')
print(f"Числовые признаки: {numeric_cols}")

# %%
# Распределение числовых признаков в зависимости от целевой переменной
fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.ravel()

for i, col in enumerate(numeric_cols):
    if i < 9:  # максимум 9 графиков
        sns.histplot(data=df, x=col, hue='Exited', 
                     kde=True, ax=axes[i], palette=['#2ecc71', '#e74c3c'])
        axes[i].set_title(f'Распределение {col} по оттоку')

plt.tight_layout()
plt.show()

# %%
# Box plots для выявления выбросов
fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.ravel()

for i, col in enumerate(numeric_cols):
    if i < 9:
        sns.boxplot(data=df, x='Exited', y=col, ax=axes[i], 
                   palette=['#2ecc71', '#e74c3c'])
        axes[i].set_title(f'{col} по группам оттока')

plt.tight_layout()
plt.show()

# %%
# Корреляция с целевой переменной
correlations = df[numeric_cols + ['Exited']].corr()['Exited'].sort_values(ascending=False)
print("Корреляция признаков с оттоком:")
print(correlations)

# Визуализация корреляций
plt.figure(figsize=(10, 6))
correlations.drop('Exited').plot(kind='bar', color=['#3498db' if x > 0 else '#e74c3c' for x in correlations.drop('Exited')])
plt.title('Корреляция признаков с оттоком')
plt.xlabel('Признаки')
plt.ylabel('Корреляция')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# %% [markdown]
# ### 7. Анализ категориальных признаков

# %%
categorical_cols = ['Geography', 'Gender']
print(f"Категориальные признаки: {categorical_cols}")

# %%
# Анализ категориальных признаков
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

for i, col in enumerate(categorical_cols):
    # Count plot
    sns.countplot(data=df, x=col, hue='Exited', ax=axes[i], 
                 palette=['#2ecc71', '#e74c3c'])
    axes[i].set_title(f'Распределение {col} по оттоку')
    axes[i].tick_params(axis='x', rotation=45)
    
    # Добавим проценты
    total = len(df)
    for p in axes[i].patches:
        height = p.get_height()
        axes[i].text(p.get_x() + p.get_width()/2., height + 5,
                f'{height/total*100:.1f}%', ha="center")

plt.tight_layout()
plt.show()

# %%
# Кросс-таблицы для категориальных признаков
for col in categorical_cols:
    print(f"\nКросс-таблица для {col}:")
    cross_tab = pd.crosstab(df[col], df['Exited'], margins=True)
    print(cross_tab)
    print(f"\nПроценты:")
    cross_tab_percent = pd.crosstab(df[col], df['Exited'], normalize='index') * 100
    print(cross_tab_percent)

# %% [markdown]
# ### 8. Анализ взаимосвязей признаков

# %%
# Матрица корреляций
plt.figure(figsize=(12, 8))
correlation_matrix = df[numeric_cols + ['Exited']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0,
            square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Матрица корреляций')
plt.tight_layout()
plt.show()

# %%
# Pairplot для ключевых признаков
key_features = ['Age', 'Balance', 'CreditScore', 'NumOfProducts', 'Exited']
sns.pairplot(df[key_features], hue='Exited', palette=['#2ecc71', '#e74c3c'], diag_kind='hist')
plt.suptitle('Pairplot ключевых признаков', y=1.02)
plt.show()

# %% [markdown]
# ### 9. Создание новых признаков и анализ

# %%
# Создадим несколько новых признаков для анализа
df['Balance_per_Product'] = df['Balance'] / (df['NumOfProducts'] + 1)  # +1 чтобы избежать деления на 0
df['Age_Tenure_Ratio'] = df['Age'] / (df['Tenure'] + 1)
df['CreditScore_per_Age'] = df['CreditScore'] / df['Age']

# Анализ новых признаков
new_features = ['Balance_per_Product', 'Age_Tenure_Ratio', 'CreditScore_per_Age']

fig, axes = plt.subplots(1, 3, figsize=(15, 5))
for i, feat in enumerate(new_features):
    sns.boxplot(data=df, x='Exited', y=feat, ax=axes[i], 
               palette=['#2ecc71', '#e74c3c'])
    axes[i].set_title(f'{feat} по группам оттока')

plt.tight_layout()
plt.show()

# %% [markdown]
# ### 10. Основные выводы EDA

# %%
print("="*50)
print("ОСНОВНЫЕ ВЫВОДЫ ИЗ EDA:")
print("="*50)

print("\n1. Дисбаланс классов:")
print(f"   - Клиентов, которые остались: {target_percent[0]:.1f}%")
print(f"   - Клиентов, которые ушли: {target_percent[1]:.1f}%")
print("   → Потребуется учитывать дисбаланс при обучении")

print("\n2. Ключевые признаки (по корреляции с оттоком):")
for feat, corr in correlations.items():
    if abs(corr) > 0.05 and feat != 'Exited':
        print(f"   - {feat}: {corr:.3f}")

print("\n3. Категориальные признаки:")
print("   - Geography: клиенты из Germany уходят чаще")
print("   - Gender: женщины уходят немного чаще мужчин")

print("\n4. Пропуски в данных:")
if missing_values.sum() == 0:
    print("   ✅ Пропуски отсутствуют - данные чистые")
else:
    print(f"   ⚠️ Есть пропуски: {missing_values.sum()}")

print("\n5. Выбросы:")
print("   - В признаках CreditScore, Age, Balance есть выбросы")
print("   → Рекомендуется использовать robust scaling или winsorizing")

print("\n6. Рекомендации по моделированию:")
print("   - Использовать методы борьбы с дисбалансом (class_weight, SMOTE)")
print("   - Удалить RowNumber, CustomerId, Surname перед обучением")
print("   - Закодировать категориальные признаки (OneHotEncoding)")
print("   - Масштабировать числовые признаки")
```
